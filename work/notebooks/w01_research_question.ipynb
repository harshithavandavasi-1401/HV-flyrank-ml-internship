## 2. The question: decision, action, cost of a wrong call
## Research Question

**Decision:** Which website pages are likely to experience a significant decline in search impressions during the next evaluation period?

**Action:** SEO teams can review the flagged pages, update outdated content, improve on-page SEO, refresh keywords, or investigate technical issues before rankings decline further.

**Cost of a Wrong Call:** If the model incorrectly predicts that a page will decline (false positive), the SEO team may spend unnecessary time updating a page that did not need attention. If the model misses a page that actually declines (false negative), the website could lose valuable organic traffic, clicks, and potential conversions because no preventive action was taken.
